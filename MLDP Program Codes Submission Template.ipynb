# 3. Data Preparation

## 3.1 Data Cleaning

```python
## Clean data
# Correct a column name that has a typo and replace single quotes with underscores
df.rename(columns = {"Nacionality": "Nationality", 
                           "Mother's qualification": "Mother_qualification", 
                           "Father's qualification": "Father_qualification", 
                           "Mother's occupation": "Mother_occupation",
                           "Father's occupation": "Father_occupation", 
                           "Age at enrollment": "Age"}, inplace = True)

# Replace white space in the column names with underscore
df.columns = df.columns.str.replace(' ', '_')

# Remove the parenthesis
df.columns = df.columns.str.replace('(', '')
df.columns = df.columns.str.replace(')', '')
```

```python
# Change data types of columns that should be categorical
col = ['Marital_status', 'Application_mode', 'Application_order', 'Course',
      'Daytime/evening_attendance', 'Previous_qualification', 'Nationality',
       'Mother_qualification', 'Father_qualification', 'Mother_occupation',
       'Father_occupation', 'Displaced', 'Educational_special_needs', 'Debtor',
       'Tuition_fees_up_to_date', 'Gender', 'Scholarship_holder',
      'International', 'Target']

df[col] = df[col].astype('category')
```

```python
from scipy.stats import chi2_contingency

categorical_cols = [
    'Marital_status', 'Application_mode', 'Application_order',
    'Course','Daytime/evening_attendance', 'Previous_qualification',
    'Nationality','Mother_qualification', 'Father_qualification',
    'Mother_occupation', 'Father_occupation', 'Displaced',
    'Educational_special_needs', 'Debtor','Tuition_fees_up_to_date',
    'Gender', 'Scholarship_holder','International'
]

chi_results = []

for col in categorical_cols:
    contingency_table = pd.crosstab(df[col], df['Target'])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    chi_results.append({
        'Feature': col,
        'Chi2': chi2,
        'p_value': p
    })

chi_df = pd.DataFrame(chi_results).sort_values('p_value')
chi_df
```

```python
# Clean Target text to avoid NaN during mapping
df['Target'] = df['Target'].astype(str).str.strip()

df['Target_encoded'] = df['Target'].map({
    'Dropout': 0,
    'Enrolled': 1,
    'Graduate': 2
})

# Remove rows that still failed mapping (safety)
df = df.dropna(subset=['Target_encoded'])
df['Target_encoded'] = df['Target_encoded'].astype(int)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import make_scorer, recall_score

recall_scorer = make_scorer(recall_score, pos_label=1)

gb_param = {
    'n_estimators': [100, 200, 300, 500],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'max_depth': [2, 3, 4, 5],
    'subsample': [0.6, 0.8, 1.0],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

gb_search_bi = RandomizedSearchCV(
    estimator=GradientBoostingClassifier(random_state=2026),
    param_distributions=gb_param,
    n_iter=30,
    scoring=recall_scorer,   # prioritize dropout recall
    cv=5,
    n_jobs=-1,
    random_state=2026,
    verbose=1
)

gb_search_bi.fit(X_train_bi, y_train_bi)

tuned_gb_bi = gb_search_bi.best_estimator_
print("Best GB Params (Binary):", gb_search_bi.best_params_)
```

```python
from sklearn.metrics import balanced_accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

y_pred = tuned_gb_bi.predict(X_test_bi)
y_prob = tuned_gb_bi.predict_proba(X_test_bi)[:, 1]

print("Tuned GB (Binary)")
print("Balanced Accuracy:", round(balanced_accuracy_score(y_test_bi, y_pred), 3))
print("Precision (Dropout):", round(precision_score(y_test_bi, y_pred), 3))
print("Recall (Dropout):", round(recall_score(y_test_bi, y_pred), 3))
print("F1 (Dropout):", round(f1_score(y_test_bi, y_pred), 3))
print("ROC-AUC:", round(roc_auc_score(y_test_bi, y_prob), 3))
```

```python
import pandas as pd
from sklearn.metrics import balanced_accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def eval_binary(model, X_test, y_test, name):
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]
    return {
        "Model": name,
        "Balanced Accuracy": balanced_accuracy_score(y_test, y_pred),
        "Precision (Dropout)": precision_score(y_test, y_pred),
        "Recall (Dropout)": recall_score(y_test, y_pred),
        "F1 (Dropout)": f1_score(y_test, y_pred),
        "ROC-AUC": roc_auc_score(y_test, y_prob)
    }

tuned_binary_df = pd.DataFrame([
    eval_binary(tuned_rf_bi, X_test_bi, y_test_bi, "Random Forest (Tuned Binary)"),
    eval_binary(tuned_gb_bi, X_test_bi, y_test_bi, "Gradient Boosting (Tuned Binary)")
]).round(3)

tuned_binary_df
```

```python
from sklearn.inspection import permutation_importance
import pandas as pd

perm = permutation_importance(
    tuned_gb_bi,
    X_test_bi,
    y_test_bi,
    n_repeats=10,
    random_state=2026,
    scoring='recall'
)

imp_df = pd.DataFrame({
    'feature': X_test_bi.columns,
    'importance': perm.importances_mean
}).sort_values('importance', ascending=False)

top_features = imp_df['feature'].head(15).tolist()
top_features
```
