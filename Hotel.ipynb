```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, roc_auc_score, classification_report

# =========================
# ITERATION 2: Feature Engineering + Feature Selection (via importance)
# =========================

# 0) (Optional) If you have a clean df already, skip.
# Make sure df is your current cleaned dataframe and includes 'is_canceled'

# 1) Feature Engineering
df = df.copy()

# total nights
if 'stays_in_week_nights' in df.columns and 'stays_in_weekend_nights' in df.columns:
    df['total_nights'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']

# binary "has special request"
if 'total_of_special_requests' in df.columns:
    df['has_special_request'] = (df['total_of_special_requests'] > 0).astype(int)

# lead time bucket
if 'lead_time' in df.columns:
    df['lead_time_group'] = pd.cut(
        df['lead_time'],
        bins=[-1, 30, 90, 180, 365, 1000],
        labels=['very_short', 'short', 'medium', 'long', 'very_long']
    )

# 2) Drop redundant/leakage columns (adjust if already dropped)
drop_cols = []

# redundant after creating total_nights
for c in ['stays_in_week_nights', 'stays_in_weekend_nights']:
    if c in df.columns:
        drop_cols.append(c)

# low value / constant-ish
if 'arrival_date_year' in df.columns:
    drop_cols.append('arrival_date_year')

# leakage (drop if still present)
for c in ['reservation_status', 'reservation_status_date']:
    if c in df.columns:
        drop_cols.append(c)

# drop safely
if drop_cols:
    df = df.drop(columns=list(set(drop_cols)))

# 3) Build X, y (One-hot encode categoricals)
y = df['is_canceled']
X = df.drop('is_canceled', axis=1)

X = pd.get_dummies(X, drop_first=True)

# 4) Train-test split (stratify is important for imbalance)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=2026, stratify=y
)

# 5) Train Random Forest (Iteration 2)
rf2 = RandomForestClassifier(
    n_estimators=200,
    criterion='entropy',
    random_state=2026,
    n_jobs=-1
)
rf2.fit(X_train, y_train)

# 6) Predict + Evaluate
y_pred = rf2.predict(X_test)
y_proba = rf2.predict_proba(X_test)[:, 1]

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nF1 Score:", f1_score(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_proba))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# 7) Feature importance (NOW valid because rf2 trained on this X)
importances = pd.Series(rf2.feature_importances_, index=X.columns).sort_values(ascending=False)

print("\nTop 20 Feature Importances:")
print(importances.head(20))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier

param_dist = {
    "n_estimators": [100, 200],          # smaller
    "max_depth": [None, 15, 25],         # fewer
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2],
    "max_features": ["sqrt", "log2"],
    "class_weight": [None, "balanced"]
}

rf = RandomForestClassifier(random_state=2026, criterion="entropy", n_jobs=-1)

search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=8,            # 🔥 cut from 20 → 8
    scoring="f1",
    cv=3,
    random_state=2026,
    n_jobs=-1,
    verbose=1
)

search.fit(X_train, y_train)

best_rf = search.best_estimator_
print("Best Params:", search.best_params_)
print("Best CV F1:", search.best_score_)
```
